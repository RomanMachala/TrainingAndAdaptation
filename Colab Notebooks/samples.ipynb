#Audio samples
In this section several audio samples are presented. This notebook was created as a part of a Bachelor's Thesis.

```python
import IPython
from google.colab import drive
import os
# imports

drive.mount('/content/drive')   # Mounts google drive
DATASETS = '/content/drive/MyDrive/thesis_training_and_adaptation/output'
# Change this part to your dataset path

def load_audios(num_of_samples: int, model: str) -> None:
  """
    Loads selected number of audios. Audios should be named sample_0X.wav - sample_0XX.wav

    Params:
      num_of_samples:     number of samples to be loaded
      model:              name of the model used for synthesis
  """
  if num_of_samples >= 100 or num_of_samples <= 0:
    print("Invalid number of samples")
    return

  data_path = os.path.join(DATASETS, model)
  for i in range(num_of_samples):
    filename = f'sample_0{i}.wav'
    audio = IPython.display.Audio(os.path.join(data_path, filename))
    IPython.display.display(audio)
```
